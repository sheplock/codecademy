```python
import pandas as pd

data = pd.read_csv('all_data.csv')
data.rename({'Life expectancy at birth (years)': 'LifeExpectancy'}, axis='columns', inplace=True)
print(data.head())
```

```python
print(data.describe())
```

```python
print(data.groupby('Country').count())
print(data.groupby('Year').count())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.hist(data.Year, bins=np.arange(2000,2017,1))
plt.show()
plt.clf()
plt.hist(data.LifeExpectancy)
plt.xlabel('Life Expectancy [years]')
plt.show()
plt.clf()
plt.subplot(1, 2, 1)
plt.hist(data.GDP/1e9)
plt.xlabel('GDP [Trillions of dollars]')
plt.subplot(1, 2, 2)
plt.hist(data.GDP/1e9, bins=[10**x for x in range(-1, 6, 1)])
plt.xlabel('GDP [Trillions of dollars]')
plt.xscale('log')
plt.show()
plt.clf()
```

The data show the life expectancy and GDP for 6 different countries over the course of 16 years from 2000-2015. The countries are: Chile, China, Germany, Mexico, USA, and Zimbabwe. In general, life expectancies range from 70-81, but there is a low tail going down to 44. GDP is similarly skewed right, where the modal GDP is in the thousands of trillions of dollars, but the lowest GDP value is just 4 trillion dollars.

```python
sns.scatterplot(x='Year', y='LifeExpectancy', hue='Country', data=data)
plt.show()
plt.clf()
```

```python
sns.scatterplot(x='Year', y='GDP', hue='Country', data=data)
plt.yscale('log')
plt.show()
plt.clf()
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,15), sharex='col')
axes[0,0].scatter(x=data.Year[data.Country=='Chile'], y=data.GDP[data.Country=='Chile'])
axes[0,0].set_title('Chile')
axes[0,1].scatter(x=data.Year[data.Country=='China'], y=data.GDP[data.Country=='China'])
axes[0,1].set_title('China')
axes[1,0].scatter(x=data.Year[data.Country=='Germany'], y=data.GDP[data.Country=='Germany'])
axes[1,0].set_title('Germany')
axes[1,1].scatter(x=data.Year[data.Country=='Mexico'], y=data.GDP[data.Country=='Mexico'])
axes[1,1].set_title('Mexico')
axes[2,0].scatter(x=data.Year[data.Country=='United States of America'], y=data.GDP[data.Country=='United States of America'])
axes[2,0].set_title('USA')
axes[2,1].scatter(x=data.Year[data.Country=='Zimbabwe'], y=data.GDP[data.Country=='Zimbabwe'])
axes[2,1].set_title('Zimbabwe')
plt.setp(axes, xticks=range(2000,2015,2))
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
fig.suptitle("GDP over Time for 6 countries", fontsize=16)
plt.xlabel("Year", fontsize=16, labelpad=10)
plt.ylabel("GDP (USD)", fontsize=16, labelpad=10)
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
plt.clf()
```

Generally, GDPs have increased over time for these 6 countries; however, there are some features within the growth. Chile's GDP shrunk from 2013-2015. Germany had lots of volatility in annual GDP from 2007-2015. Mexico had a few down years - 2009, 2010, 2015. The United States' growth slowed in 2008, GDP shrunk in 2009, but has been increasing since. Zimbabwe had a period of decline from 2000 to 2008 not seen in the other countries before fast growth until 2012 when GDP increases slowed. The only country in this sample that experienced growth only from 2000 to 2015 was China.

```python
sns.scatterplot(x='Year', y='LifeExpectancy', hue='Country', data=data)
plt.show()
plt.clf()
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,15), sharex='col')
axes[0,0].scatter(x=data.Year[data.Country=='Chile'], y=data.LifeExpectancy[data.Country=='Chile'])
axes[0,0].set_title('Chile')
axes[0,1].scatter(x=data.Year[data.Country=='China'], y=data.LifeExpectancy[data.Country=='China'])
axes[0,1].set_title('China')
axes[1,0].scatter(x=data.Year[data.Country=='Germany'], y=data.LifeExpectancy[data.Country=='Germany'])
axes[1,0].set_title('Germany')
axes[1,1].scatter(x=data.Year[data.Country=='Mexico'], y=data.LifeExpectancy[data.Country=='Mexico'])
axes[1,1].set_title('Mexico')
axes[2,0].scatter(x=data.Year[data.Country=='United States of America'], y=data.LifeExpectancy[data.Country=='United States of America'])
axes[2,0].set_title('USA')
axes[2,1].scatter(x=data.Year[data.Country=='Zimbabwe'], y=data.LifeExpectancy[data.Country=='Zimbabwe'])
axes[2,1].set_title('Zimbabwe')
plt.setp(axes, xticks=range(2000,2015,2))
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
fig.suptitle("Life Expectancy over Time for 6 countries", fontsize=16)
plt.xlabel("Year", fontsize=16, labelpad=10)
plt.ylabel("Life Expectancy (Years)", fontsize=16, labelpad=10)
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
plt.clf()
```

Similar to GDP, life expectancy has increaesd over time with a few exceptions. Zimbabwe saw a decrease from 2000 to 2004 before a steady increase in life expectancy, up 15 years on average from 2004 to 2015. Mexico, Germany, and Chile have had periods of a few years of statis or decrease within a general increase over time. The United States and China have generally increased every year, though the total increse over this period for the US is about 3 years, while China's total increase was 4.5 years. Among these countries, Germany had the highest 2015 life expectancy, around 81 years.

```python
sns.scatterplot(x='GDP', y='LifeExpectancy', hue='Country', data=data)
plt.xscale('log')
plt.show()
plt.clf()
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,15))
axes[0,0].scatter(x=data.GDP[data.Country=='Chile'], y=data.LifeExpectancy[data.Country=='Chile'])
axes[0,0].set_title('Chile')
axes[0,1].scatter(x=data.GDP[data.Country=='China'], y=data.LifeExpectancy[data.Country=='China'])
axes[0,1].set_title('China')
axes[1,0].scatter(x=data.GDP[data.Country=='Germany'], y=data.LifeExpectancy[data.Country=='Germany'])
axes[1,0].set_title('Germany')
axes[1,1].scatter(x=data.GDP[data.Country=='Mexico'], y=data.LifeExpectancy[data.Country=='Mexico'])
axes[1,1].set_title('Mexico')
axes[2,0].scatter(x=data.GDP[data.Country=='United States of America'], y=data.LifeExpectancy[data.Country=='United States of America'])
axes[2,0].set_title('USA')
axes[2,1].scatter(x=data.GDP[data.Country=='Zimbabwe'], y=data.LifeExpectancy[data.Country=='Zimbabwe'])
axes[2,1].set_title('Zimbabwe')
#plt.setp(axes, xticks=range(2000,2015,2))
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
fig.suptitle("Life Expectancy vs GDP for 6 countries", fontsize=16)
plt.xlabel("GDP (USD)", fontsize=16, labelpad=10)
plt.ylabel("Life Expectancy (Years)", fontsize=16, labelpad=10)
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
plt.clf()
```

Just as GDP and Life Expectancy have increased over time, the two measures are correlated, although not perfectly and not always linearly. In the Zimbabwe data, GDP increases from about 4 trillion to 700 trillion dollars didn't always come with an increase in life expectancy, but above 800 trillion dollars, there is a more clear linear relationship. Chile, Germany, and Mexico don't have quite as strong a linear relationship as the USA measures do. Finally, China's relationship looks almost logarithmic, with large life expectancy increases with small GDP increases at the lower end of their GDP in this sample, and then at higher GDP, life expectancy still increases with GDP but at a slower rate.
